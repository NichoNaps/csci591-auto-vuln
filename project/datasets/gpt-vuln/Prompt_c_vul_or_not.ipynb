```python
import pandas as pd
import random

train_data_path = 'F:/vuldetection/cleaned_train_data.csv'
validation_data_path = 'F:/vuldetection/Cleaned_validation_for_codexglue_binary.csv'
test_data_path = 'F:/vuldetection/Cleaned_test_for_codexglue_binary (1).csv'

# Prompt generation parameters
limit = True  # Set to False to generate all prompts 
prompt_limit = 10  # Maximum number of prompts if limit is True

# Define variations for each step
intro_variants = [
    "Analyze the following code snippet for vulnerability detection.",
    "Evaluate this C code snippet and decide if it has potential vulnerabilities.",
]

step1_variants = [
    "Describe the primary purpose or function of this code.",
    "Explain what the main function of this code seems to be."
]

step2_variants = [
    "Identify any patterns or operations that might introduce errors or risks.",
    "Check for any risky operations or patterns that could lead to potential vulnerabilities."
]

step3_variants = [
    "Does this code include any validation or error handling mechanisms?",
    "Note if there are any safeguards, such as input checks or validations."
]

step4_variants = [
    "Conclude if the code is vulnerable (1) or not (0). Summarize your reasoning.",
    "Based on your analysis, classify the code as vulnerable (1) or secure (0) and explain why."
]

# Function to generate prompts for a specific dataset
def generate_prompts(dataset, output_file):
    prompts = []
    
    # Prompt generation loop
    for index, row in dataset.iterrows():
        project_name = row.get("project", "Unknown Project")
        code_snippet = row.get("func", "")
        
        # Randomly select a variant for each part
        intro = random.choice(intro_variants)
        step1 = random.choice(step1_variants)
        step2 = random.choice(step2_variants)
        step3 = random.choice(step3_variants)
        step4 = random.choice(step4_variants)
        
        # Construct the prompt
        prompt = f"""
        {intro} Here is a code snippet from the project '{project_name}'.

        Code Snippet:
        {code_snippet}

        Step 1: {step1}
        Step 2: {step2}
        Step 3: {step3}
        Step 4: {step4}
        """
        
        prompts.append(prompt)
        
        if limit and len(prompts) >= prompt_limit:
            break

    # Save prompts to file
    with open(output_file, "w") as f:
        for prompt in prompts:
            f.write(prompt + "\n\n")
    print(f"Prompts generated and saved to '{output_file}'")

# Load datasets
train_data = pd.read_csv(train_data_path)
validation_data = pd.read_csv(validation_data_path)
test_data = pd.read_csv(test_data_path)

# prompts for each dataset separately
generate_prompts(train_data, "F:/vuldetection/prompts_train_data.txt")
generate_prompts(validation_data, "F:/vuldetection/prompts_validation_data.txt")
generate_prompts(test_data, "F:/vuldetection/prompts_test_data.txt")
```
